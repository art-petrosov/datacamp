## 10. if, elif, else

### Overview

* Comparison Operators
    * `<`, `>`, `>=`, `<=`, `==`, `!=`
* Boolean Operators
    * `and`, `or`, `not`
* Conditional Statements
    * `if`, `else`, `elif`

### `if`

````python
z = 4

if z % 2 == 0 :
    print("z is even")
````

````python
z = 4

if z % 2 == 0 :
    print("checking " + str(z))
    print("z is even")
````

### `else`

````python
z = 5

if z % 2 == 0 :
    print("z is even")
else:
    print('z is odd')
````

### `elif`

````python
z = 3

if z % 2 == 0 :
    print("z is divisible by 2")
elif z % 3 == 0 :
    print("z is divisible by 3")
else :
    print("z is neither divisible by 2 nor by 3")
````

````python
np_height = np.array([1.73, 1.68, 1.71, 1.89, 1.79])
np_weight = np.array([65.4, 59.2, 63.6, 88.4, 68.7])
meas = np.array([np_height, np_weight])
````

## 16. Random Walk

### Heads or Tails

* 0: heads
* 1: tails

````python
np.random.seed(123)
outcomes = []

for x in range(10) :
    coin = np.random.randint(0, 2)
    
    if coin == 0:
        outcomes.append("heads")
    else:
        outcomes.append("tails")
    
print(outcomes)
````

### Heads or Tails: Random Walk

````python
np.random.seed(123)
tails = [0]

for x in range(10) :
    coin = np.random.randint(0, 2)
    tails.append(tails[x] + coin)
    
print(tails)
````

### Step to Walk

* outcomes  
`['heads', 'tails', 'heads', 'heads', 'heads', 'heads', 'heads', 'tails', 'tails', 'heads']`

* tails  
`[0, 0, 1, 1, 1, 1, 1, 1, 2, 3, 3]`

## 17. Distribution

### 100 runs

````python
np.random.seed(123)
final_tails = []

for x in range(100) :
    tails = [0]
    
    for x in range(10) :
        coin = np.random.randint(0,2)
        tails.append(tails[x] + coin)
        
    final_tails.append(tails[-1])
        
print(final_tails)
````

### Histogram, 100 runs

````python
plt.hist(final_tails, bins = 10)
plt.show()
````

### Histogram, 1.000 runs

````python
np.random.seed(123)
final_tails = []

for x in range(1000) :
    tails = [0]
    
    for x in range(10) :
        coin = np.random.randint(0,2)
        tails.append(tails[x] + coin)
        
    final_tails.append(tails[-1])
        
plt.hist(final_tails, bins=10)
plt.show()
````
